```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array ([0,10])
y=np.array([20,100])
print(plt.plot(x,y))
print(plt.show)
```

```python
x=np.array([0,5])
y=np.array([0,25])
print(plt.plot(x,y,marker='p'))
print(plt.show)
```
